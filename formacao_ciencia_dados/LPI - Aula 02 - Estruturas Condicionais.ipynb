### 6. Expressões aritméticas

Como podemos observar no exemplo anterior, o Python faz operações aritméticas de maneira bastante intuitiva, similar ao que estamos acostumados. Os operadores aceitos são:

* Soma: +

* Subtração: -
 
* Multiplicação: *

* Divisão: /

* Divisão inteira: //

* Resto da divisão: %

* Potência: **



```python
numero1 = int(input('Digite um número: '))
numero2 = int(input('Digite outro número:'))
soma = numero1 + numero2
subtracao = numero1 - numero2
multiplicacao = numero1 * numero2
divisao_real = numero1 / numero2
divisao_inteira = numero1 // numero2
resto = numero1 % numero2
elevado = numero1 ** numero2
print('Soma: ', soma)
print('Subtração: ', subtracao)
print('Multiplicação: ', multiplicacao)
print('Divisão: ', divisao_real)
print('Divisão inteira: ', divisao_inteira)
print('Resto da divisão: ', resto)
print('Potência: ', elevado)
```

Imagine que o critério de aprovação em uma escola seja a média superior a 7.0 e presença superior a 75%. Neste caso, o aluno precisa atender a ambos os critérios para ser aprovado. Se ele tirou uma ótima nota, mas faltou demais, será reprovado. Se ele compareceu a todas as aulas, mas teve notas baixas, idem.

```python
media = float(input('Digite a média do aluno: '))
presenca = float(input('Digite as presenças do aluno: '))

aprovado = media >= 7.0 and presenca >= 0.75
print('O aluno foi aprovado?', aprovado)
```

## Estrutura condicional

Em nossos códigos, usualmente queremos que, dependendo dos valores das variáveis, a gente tenha respostas diferentes.
Em Python, faremos isso com o auxílio de três nomes especiais: `if`, `else` e `elif`(else if).

Estes comandos funcionam da seguinte forma:

```python
# if condicao:
#   faz algo
```

Perceba que, quando usamos o `if`, precisamos nos atentar em algumas novidades:
- Precisamos definir uma condição. Essa condição deve retornar um valor booleano. Isto é, dizer se a condição é verdadeira ou não. Caso a condição seja verdadeira, o código realiza o que está dentro deste bloco. Caso contrário, ele apenas "ignora" o que está dentro.

- Precisamos usar o `:`. 

- Precisamos definir o que acontece se a condição é verdadeira.

```python
nota1 = float(input('Digite a nota1: '))
nota2 = float(input('Digite a nota2: '))

media = (nota1 + nota2) / 2

if media >= 7.0: ## if -> comando condicional
    print('Aprovado')

print('Média: ', media)
```

E se quisermos escrever também se a pessoa estiver reprovada?

Neste caso, usar o `else` faz muito mais sentido do que usar dois `if`. Intuitivamente, o que é mais fácil?

1. Checar se a média é maior que 7 e, se não for verdadeiro, dizer que está reprovado.

2. Checar se a média é maior que 7 para dizer se está aprovado, e depois checar novamente se a média é menor que 7 para dizer se está reprovado.

E se quisermos adicionar mais uma condição para a nossa aprovação? Por exemplo, se `nota` for maior que 7 e `presenca` maior que 0.75, então está aprovado.

Vamos incluir a Prova Final nesse exemplo. Com isso, temos a seguinte condição:

- Se o aluno tem nota > 7 - Aprovado.
- Caso contrário, se o aluno tem nota > 5 - Prova Final.
- Caso contrário - Reprovado.

```python
nota1 = float(input('Digite a nota1: '))
nota2 = float(input('Digite a nota2: '))

media = (nota1 + nota2) / 2
```

```python
x = int(input("Digite o número x: "))
y = int(input("Digite o número y: "))

if x % y == 0:
    print(f"{x} é divisível por {y}")
    resultado = x // y 
    print(f"Resultado da divisão de {x} por {y} é {resultado}")
else:
    print(f"{x} não é divisível por {y}")
```
